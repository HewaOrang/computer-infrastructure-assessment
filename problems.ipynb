# The download() functions

```python
# Get data.
df = yf.download(['MSFT', 'AAPL', 'GOOG'], period='5d', interval='1h')

# Show.
df
```

```python
# The index.
df.index
```

```python
# print data as csv.
# https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.to_csv.html
print(df.to_csv())
```

```python
# Save data as csv.
# https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.to_csv.html
df.to_csv('../data/data.csv')
```
